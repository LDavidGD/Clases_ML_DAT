Prueba de Normalidad de Shapiro-Wilk

```python
# Al hacer una prueba de hipÃ³tesis debe darse una variable de confianza \alpha
```

```python
from scipy.stats import shapiro
```

```python
estadistico, p_value = shapiro(df['CO'])
print('Valor del estadÃ­stico = {%.3f} y p_value = %.3f' % (estadistico, p_value))
```

```python
# p_value < 0.05 por lo que se acepta la hipÃ³tesis alternativa. Es decir, no es una distribuciÃ³n normal!
```

# Factores meteorolÃ³gicos

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/jamc88/Analisis-datos-atmosfericos/refs/heads/main/datos/redmet_2023_05.csv",
                 parse_dates = True, index_col=0)
df2
```

```python
estadistico, p_value = shapiro(df2['RH'])
print('valor estadistico: %.3f, p_value = %.3f' % (estadistico, p_value))
```

```python
cm = pd.concat([df, df2], axis = 1)
cm
```

```python
correlation_matrix2 = cm.corr('spearman')
plt.figure(figsize = (8,6))
sns.heatmap(correlation_matrix2, annot = True, cmap = 'coolwarm', vmin = -1, vmax = 1, center = 0)
```

```python
# Repitiendo el proceso con el Penguin, pero para cm

corr = pg.pairwise_corr(cm, method = 'spearman')
c = corr.sort_values(by = 'r', ascending = False)[['X', 'Y', 'r']]
c = c[(c['r'] > 0.5) | (c['r'] < -0.5)]
c
```

```python
sns.scatterplot(x = 'RH', y = 'PM10', data = cm)
```
